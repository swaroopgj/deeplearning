Loading cropped Yale faces and rescale by 1/3. for VAE

```python
from skimage.io import imread
from skimage.transform import rescale
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import glob
```

```python
fnames = sorted(glob.glob('CroppedYale/yaleB??/yaleB??_P00A*.pgm'))
```

```python
faces = [imread(f) for f in fnames]
```

```python
scale = 1/3.0
faces_scaled = [rescale(face, scale, cval=128) for face in faces]
```

```python
for i in range(1,65):
    plt.subplot(8,8,i);
    plt.imshow(faces_scaled[i], cmap=plt.cm.Greys_r)
    plt.axis('off')
```

```python
# only use faces with at least mean=0.2 to remove too-dark illumination
lit_faces_scaled = [face for face in faces_scaled if face.mean()>0.2]
```

```python
import pickle
```

```python
with open('lit_scaled_facse.pkl', 'wb') as f:
    pickle.dump(lit_faces_scaled, f)
```
